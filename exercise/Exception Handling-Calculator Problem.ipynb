# Calculator Program Using Exception Handling

```python
class FormulaError(Exception):
    pass
```

```python
while True:
    cal=input().split(" ")
    if cal[0]=="quit":
        break
    elif len(cal)<3 or len(cal)>3:
        raise FormulaError("Enter the Equation in correct Length")
    elif cal[1] not in ["+","-","*","/"]:
        raise FormulaError("Operator Mismatch: Only + - * / are Available for calculation")
    elif not(cal[0].isdigit()) or not(cal[2].isdigit()):
        raise FormulaError("Enter the value in digits")
    out="".join(cal)
    print(eval(out))
```
