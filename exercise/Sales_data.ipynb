```python
import pandas as pd
url="https://raw.githubusercontent.com/svkarthik86/Assignment/main/Sales_Data/Sales_January_2019.csv"
sales_data=pd.read_csv(url)
sales_data.head(4)
```

```python
sales_data.columns=sales_data.columns.str.replace(" ","_")
sales_data.dropna(inplace=True)
sales_data
```

```python
sales_data.drop_duplicates(inplace=True)
sales_data[sales_data.duplicated()]
sales_data=sales_data[~(sales_data.Price_Each=="Price Each")]
sales_data
```

```python
sales_data["order_price"]=sales_data.Price_Each.astype(float)*sales_data.Quantity_Ordered.astype(float)
sales_data
```

```python
c=sales_data.drop('Purchase_Address',1)
c
```

```python
sales_data=pd.concat((sales_data,pd.DataFrame([[100001,"OnePhone",1,600,"01/22/19 21:25",600]],columns=['Order_ID', 'Product', 'Quantity_Ordered', 'Price_Each', 'Order_Date',
       'order_price'])),ignore_index=True)
sales_data
```

```python
import numpy as np
sales_data.groupby("Product")["order_price"].agg([min,max,np.mean])
```
